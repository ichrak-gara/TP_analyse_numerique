```python
def dicho(f,a,b,tol):
    m=(a+b)/2
    err=abs(b-a)
    while err>tol :
        if f(m)==0 :
            break
   
        if f(a)*f(m)<0 :
            b=m
        else: a=m
    
        m=(a+b)/2
        err=abs(b-a)
    return m
```

```python
print(dicho(g,-2,2,0.001))
```

```python
def dichow(f,a,b,tol):
    m=(a+b)/2
    err=abs(b-a)
    i=0
    while err>tol :
        if f(m)==0 :
            break
   
        if f(a)*f(m)<0 :
            b=m
        else: a=m
    
        m=(a+b)/2
        err=abs(b-a)
        i=i+1
    return (m,i)
```

```python
print(dichow(g,-2,-1.5,0.0001))
```

```python
def newton(a,eps):
    x1=a
    x2=x1-f(x1)/fprime(x1)
    
    while abs(x1-x2)>eps:
        x1=x2
        x2=x1-f(x1)/fprime(x1)
        sol=x2
    return sol
```
